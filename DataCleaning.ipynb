```python
dataset_df = pd.DataFrame(dataset) #to dataframe
```

```python
# Melt the DataFrame
value_vars = dataset_df.columns.difference(['Unnamed: 0', 'year']).tolist()

df_long = pd.melt(dataset_df, id_vars=['Unnamed: 0','country','year'],value_vars=value_vars,
 var_name='key', value_name='cases')

# Drop rows with NA values in the 'cases' column
df_long = df_long.dropna(subset=['cases'])
```

**Replace newrel with "new_rel" from key column**

```python

df_long['key'] = df_long['key'].str.replace('newrel', 'new_rel')
```

```python
split = df_long['key'].str.split("_", n=2, expand=True)
```

```python
df_long["new"] = split[0]
df_long["type"] = split[1]
df_long["sexage"] = split[2]
```

```python
df_long.head(5)
```

**Drop column "new" and split sexage column also**

```python
df_long = df_long.drop(['new'], axis=1)
```

```python
df_long[['sex', 'age']] = df_long['sexage'].str.extract('([a-zA-Z]+)([0-9]+)')
```

```python
df_long.head(5)
```

```python
df_long = df_long.drop(['sexage','Unnamed: 0'], axis=1)
```

```python
df_long = df_long.drop(['key'], axis=1)
```

```python
def MissingValues():
  country = df_long['country'].isna().sum()
  year = df_long['year'].isna().sum()
  cases = df_long['cases'].isna().sum()
  types = df_long['type'].isna().sum()
  sex = df_long['sex'].isna().sum()
  age = df_long['age'].isna().sum()
  print(country)
  print("Missing Values")
  print(country,year,cases,types,sex,age)
MissingValues()
```

```python
df_long['age'] = df_long['age'].astype('int')
```

```python
df_long['age'].nunique()
```

```python
def replace_values(value):
    if value == 14:
        return '0-14'
    elif value == 4554:
        return '45-54'
    elif value == 3544:
        return '35-44'
    elif value == 1524:
        return '15-24'
    elif value == 5564:
        return '55-64'
    elif value == 2534:
        return '25-34'
    else:
        return '65-above'

# Apply the custom function to the specified column
df_long['age'] = df_long['age'].apply(replace_values)
```
